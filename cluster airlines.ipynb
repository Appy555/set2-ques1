```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram
import seaborn as sn
```

```python
Air = pd.read_csv('F://EastWestAirlines.csv')
Air
```

```python
Air.info()
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(Air.iloc[:,1:])
df_norm
```

```python
plt.figure(figsize=(10,7))
dendrogram = sch.dendrogram(sch.linkage(df_norm, method = 'ward'))
```

```python
hclusters = AgglomerativeClustering(n_clusters = 5, affinity='euclidean', linkage='ward')
hclusters
```

```python
y = pd.DataFrame(hclusters.fit_predict(df_norm), columns=['clusterid'])
y['clusterid'].value_counts()
```

```python
Air['clusterid']=hclusters.labels_
Air
```

```python
Air.groupby('clusterid').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(Air['clusterid'],Air['Balance'], c=hclusters.labels_)

```

# KMeans

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
from scipy.spatial.distance import cdist
```

```python
Airl = pd.read_csv('F://EastWestAirlines.csv')
```

```python
Airl
```

```python
df_norm=norm_func(Airl.iloc[:,1:])
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',random_state = 0)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
    plt.plot(range(1,11),wcss)
    plt.title('Elbow Curve')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS')
    plt.show()
    
    
```

```python
wcss
```

```python
from sklearn.cluster import KMeans
cluster_new = KMeans(n_clusters = 4)
cluster_new.fit(df_norm)
cluster_new.labels_
```

```python
Airl.iloc[:,1:5].groupby(Airl.Cluster_new).mean()
```

# DBSCAN


```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('F://EastWestAirlines.csv')
print(df.head())
```

```python
print(df.info())
```

```python

df = df.iloc[:,1:5]
```

```python
df.values
```

```python
stscaler = StandardScaler().fit(df.values)
x = stscaler.transform(df.values)
```

```python
x
```

```python
dbscan = DBSCAN(eps = 2, min_samples = 5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_, columns = ['cluster'])
cl
```

```python
pd.concat([df,cl], axis = 1)
```
