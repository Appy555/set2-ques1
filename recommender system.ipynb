```python
data1 = data.pivot_table(index="User.ID", columns="Book.Title", values="Book.Rating", aggfunc="sum")
data1
```

```python
data1.index = data.UserID.unique()
data1
```

```python
data1.fillna(0,inplace=True)
data1
```

```python
# Calculating Cosine Similarity between users :
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
data_user = 1-pairwise_distances(data1.values,metric="cosine")
data_user
```

```python
data_user_df = pd.DataFrame(data_user)
```

```python
data_user_df.index = data.UserID.unique()
data_user_df.columns = data.UserID.unique()
```
